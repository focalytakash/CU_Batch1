```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/anna/Downloads/canada_per_capita_income.csv')
df.rename(columns={'per capita income (US$)':'income'},inplace=True)
df
```

```python
plt.scatter(df.year,df.income)
```

```python
x = df[['year']]
y = df['income']
```

```python
from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression

lg = LinearRegression()

lg.fit(x_train,y_train)
```

```python
print(x_test)
lg.predict(x_test)
```

```python
y_test
```

```python
lg.score(x_test,y_test)
```

```python
# Logistic regression

df = pd.read_csv('/Applications/Aquib Folder/Excel Full Class/placement.csv')

x = df[['cgpa']]
y = df['placed']
```
